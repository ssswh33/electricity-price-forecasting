### Creating new features

```python
season = []
spring= ['FEBRUARY','MARCH','APRIL','MAY']
summer = ['JUNE','JULY','AUGUST']
fall = ['SEPTEMBER','OCTOBER','NOVEMBER']
winter = ['DECEMBER','JANUARY']
for x in data['MONTH'].values:
    if x in spring:
        season.append('1')
    elif x in summer:
        season.append('2')
    elif x in fall:
        season.append('3')
    else:
        season.append('4')
```

```python
data['SEASON'] = season
```

### Label Encoding for categorical features

```python
peaktype = ['OFFPEAK','ONPEAK']
off_idx = data[data['PEAKTYPE'] == 'OFFPEAK'].index
on_idx = data[data['PEAKTYPE'] == 'ONPEAK'].index
peaktype_idx = data.columns.get_loc('PEAKTYPE')
for i in off_idx:
    data.iloc[i,peaktype_idx] = 0
for i in on_idx:
    data.iloc[i,peaktype_idx] = 1
```

```python
month = list(data.MONTH.unique())

month_idx = data.columns.get_loc('MONTH')
for x in month:
    idx = data[data['MONTH'] == x].index
    for i in idx:
        data.iloc[i,month_idx] = int(month.index(x)) + 1
```
